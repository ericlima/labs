```
%scala

import org.apache.spark.sql.avro.functions.from_avro
import org.apache.avro.SchemaBuilder
```

> https://spark.apache.org/docs/latest/sql-data-sources-avro.html

```
%scala

//kafka
val kafkaBrokers = "20.72.126.44:9094"

//schema registry
val schemaRegistryAddr = "http://20.85.55.31:8081"

//topic
val rides_topic = "src-app-rides-avro"

//avro
val rides_avro_key = "src-app-rides-avro-key"
val rides_avro_value = "src-app-rides-avro-value"
```

```
%scala

//kafka
val df_kafka_mssql = spark
  .readStream
  .format("kafka")
  .option("kafka.bootstrap.servers", kafkaBrokers)
  .option("subscribe", rides_topic)
  .option("startingOffsets", "earliest")
  .load()
  .select(from_avro($"key", rides_avro_key, schemaRegistryAddr).as("key"), from_avro($"value", rides_avro_value, schemaRegistryAddr).as("value"))

// is streaming ? 
df_kafka_mssql.isStreaming
```

```
%scala

df_kafka_mssql 
  .writeStream
  .format("delta") 
  .outputMode("append")
  .option("checkpointLocation", "dbfs:/delta/reviews/_checkpoints/stream_kafka_bronze_reviews") 
  .start("/delta/stream_kafka_bronze_reviews")
```

```
%sql

-- 19.863
SELECT COUNT(*) AS amt
FROM delta.`/delta/stream_kafka_bronze_reviews`
```

```
%sql

SELECT value.*
FROM delta.`/delta/stream_kafka_bronze_reviews`
```
